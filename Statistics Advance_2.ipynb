## Question 1

**What is hypothesis testing in statistics?**

**Answer:**

**Hypothesis testing** is a formal procedure for investigating our ideas about the world using statistical methods. It involves a systematic approach to making an informed decision about a claim or a hypothesis about a population parameter, such as a population mean or proportion. The process involves formulating a null hypothesis and an alternative hypothesis, collecting data, analyzing it, and then deciding whether to reject the null hypothesis in favor of the alternative or to fail to reject the null hypothesis.

## Question 2

**What is the null hypothesis, and how does it differ from the alternative hypothesis?**

**Answer:**

The **null hypothesis** ($H_0$) is a statement of no effect or no difference. It represents the default assumption that there is no relationship or no difference between the variables being tested. It is the statement that the researcher is trying to disprove or reject.

The **alternative hypothesis** ($H_a$ or $H_1$) is a statement that contradicts the null hypothesis. It represents the claim or the effect that the researcher is trying to find evidence for. It proposes that there is a relationship or a difference between the variables. The goal of a hypothesis test is to determine whether the evidence is strong enough to reject the null hypothesis in favor of the alternative hypothesis.

## Question 3

**Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.**

**Answer:**

The **significance level** ($ \alpha $) is the probability of rejecting the null hypothesis when it is actually true. It is a threshold that we set before conducting a hypothesis test to determine how much evidence we need to reject the null hypothesis. Common significance levels are $0.05$ (5%), $0.01$ (1%), and $0.10$ (10%).

Its role is crucial in deciding the outcome of a test. If the p-value (the probability of observing the data given that the null hypothesis is true) is less than or equal to the significance level ($p \le \alpha$), we **reject the null hypothesis**. This means our results are statistically significant, and there is strong evidence to support the alternative hypothesis. If the p-value is greater than the significance level ($p > \alpha$), we **fail to reject the null hypothesis**. This means there is not enough evidence to conclude a statistically significant effect.

## Question 4

**What are Type I and Type II errors? Give examples of each.**

**Answer:**

In hypothesis testing, there are two types of errors we can make:

1.  **Type I Error (False Positive):** This occurs when we **reject the null hypothesis** when it is, in fact, **true**. The probability of a Type I error is equal to the significance level, $ \alpha $. 
    * **Example:** A medical test for a disease incorrectly concludes that a healthy person has the disease. The null hypothesis is that the person is healthy, and we mistakenly reject it.

2.  **Type II Error (False Negative):** This occurs when we **fail to reject the null hypothesis** when it is, in fact, **false**. The probability of a Type II error is denoted by $ \beta $. 
    * **Example:** A medical test for a disease incorrectly concludes that a sick person is healthy. The null hypothesis is that the person is healthy, and we mistakenly fail to reject it.

## Question 5

**What is the difference between a Z-test and a T-test? Explain when to use each.**

**Answer:**

Both the Z-test and the T-test are statistical hypothesis tests used to compare a sample mean to a population mean or to compare two sample means. The key difference lies in what we know about the population standard deviation ($ \sigma $).

* A **Z-test** is used when the **population standard deviation is known** or the sample size is large ($n > 30$). The test statistic follows a standard normal (Z) distribution. 
* A **T-test** is used when the **population standard deviation is unknown** and the sample size is small ($n < 30$). The test statistic follows a t-distribution, which has heavier tails than the normal distribution to account for the additional uncertainty from estimating the population standard deviation from the sample.

## Question 6

**Write a Python program to generate a binomial distribution with $n=10$ and $p=0.5$, then plot its histogram.**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Parameters for the binomial distribution
n = 10  # number of trials
p = 0.5 # probability of success
num_experiments = 1000  # number of times to run the binomial experiment

# Generate random numbers from a binomial distribution
binomial_data = np.random.binomial(n, p, num_experiments)

# Plot the histogram
plt.figure(figsize=(10, 6))
sns.histplot(binomial_data, bins=range(n + 2), kde=False, color='skyblue', discrete=True)
plt.title('Binomial Distribution ($n=10, p=0.5$)', fontsize=16)
plt.xlabel('Number of Successes', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.xticks(range(n + 1))
plt.grid(axis='y', alpha=0.75)
plt.show()
```

## Question 7

**Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.**

**sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6, 50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5, 50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9, 50.3, 50.4, 50.0, 49.7, ... ]**

```python
import numpy as np
from scipy import stats

# Given sample data
sample_data = np.array([
    49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
    50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
    50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
    50.3, 50.4, 50.0, 49.7
])

# Assume a population mean (this is our null hypothesis)
pop_mean_H0 = 50.0

# Assume we know the population standard deviation for a Z-test
# In a real-world scenario, this might be from historical data.
pop_std_dev_sigma = 1.0

# Significance level
alpha = 0.05

# Calculate sample statistics
sample_mean = np.mean(sample_data)
sample_size = len(sample_data)

# Calculate the Z-statistic
z_statistic = (sample_mean - pop_mean_H0) / (pop_std_dev_sigma / np.sqrt(sample_size))

# Calculate the p-value for a two-tailed test
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

print(f"Sample Mean: {sample_mean:.4f}")
print(f"Z-statistic: {z_statistic:.4f}")
print(f"p-value: {p_value:.4f}")
print(f"Significance Level (alpha): {alpha}")

print("\n--- Interpretation ---")
if p_value < alpha:
    print("Since the p-value is less than the significance level, we reject the null hypothesis.")
    print("There is statistically significant evidence to conclude that the sample mean is different from the population mean of 50.0.")
else:
    print("Since the p-value is greater than the significance level, we fail to reject the null hypothesis.")
    print("There is not enough statistically significant evidence to conclude that the sample mean is different from the population mean of 50.0.")
```

## Question 8

**Explain the concept of p-value and how it is used to draw conclusions in hypothesis testing.**

**Answer:**

The **p-value** is the probability of observing a test statistic as extreme as, or more extreme than, the one calculated from our sample data, assuming that the null hypothesis is true. It quantifies the strength of the evidence against the null hypothesis.

A **small p-value** (typically less than or equal to $ \alpha $) indicates that the observed data is very unlikely to have occurred if the null hypothesis were true. This provides strong evidence against the null hypothesis, leading us to **reject it**.

A **large p-value** (greater than $ \alpha $) indicates that the observed data is not particularly unusual under the null hypothesis. This means there is not enough evidence to reject the null hypothesis, so we **fail to reject it**.

## Question 9

**Write a Python program to create and visualize a normal distribution curve with a mean of 50 and standard deviation of 10.**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

# Parameters for the normal distribution
mu = 50
sigma = 10

# Create a range of x values
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)

# Calculate the probability density function (PDF) for each x value
pdf = norm.pdf(x, mu, sigma)

# Plot the normal distribution curve
plt.figure(figsize=(10, 6))
sns.lineplot(x=x, y=pdf, color='darkgreen', linewidth=2)

# Add a dashed line for the mean
plt.axvline(x=mu, color='red', linestyle='--', label=f'Mean = {mu}')

# Add titles and labels
plt.title('Normal Distribution Curve ($\\mu=50, \\sigma=10$)', fontsize=16)
plt.xlabel('Value', fontsize=12)
plt.ylabel('Probability Density', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```
