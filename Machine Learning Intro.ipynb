## Question 1

**Explain the differences between AI, ML, Deep Learning (DL), and Data Science (DS).**

**Answer:**

Here's a breakdown of the differences and how these fields relate to each other:

* **Artificial Intelligence (AI):** The broadest field. It's the science of making machines intelligent, enabling them to perform tasks that typically require human intelligence, such as problem-solving, decision-making, and understanding language.

* **Machine Learning (ML):** A subfield of AI. It focuses on the development of algorithms that allow computers to learn from data without being explicitly programmed. ML models identify patterns and make predictions based on data.

* **Deep Learning (DL):** A subfield of ML. It uses neural networks with many layers (hence "deep") to model complex patterns in data. Deep learning is particularly effective for tasks like image recognition, natural language processing, and speech recognition.

* **Data Science (DS):** An interdisciplinary field that uses scientific methods, processes, algorithms, and systems to extract knowledge and insights from data. While it uses ML and DL techniques, it also includes aspects of statistics, data visualization, and domain expertise to solve real-world problems. 

## Question 2

**What are the types of machine learning? Describe each with one real-world example.**

**Answer:**

The three primary types of machine learning are:

1.  **Supervised Learning:** The model is trained on a labeled dataset, which means both the input data and the correct output are provided. The goal is for the model to learn the mapping from inputs to outputs.
    * **Example:** Predicting house prices based on features like square footage, number of bedrooms, and location. The model is trained on a dataset where both the features (inputs) and the corresponding house prices (outputs) are known.

2.  **Unsupervised Learning:** The model is trained on an unlabeled dataset. The algorithm must find hidden patterns or intrinsic structures within the data on its own. It's used for tasks like clustering and association.
    * **Example:** A customer segmentation algorithm that groups customers into different segments based on their purchasing behavior without any prior labels.

3.  **Reinforcement Learning:** The model learns to make a sequence of decisions in an environment to maximize a reward. It's like training a pet with rewards; the agent takes actions and learns from the consequences.
    * **Example:** An AI playing a game like Chess or Go. The agent learns the best moves by receiving a positive reward for winning and a negative reward for losing.

## Question 3

**Define overfitting, underfitting, and the bias-variance tradeoff in machine learning.**

**Answer:**

1.  **Overfitting:** This happens when a model learns the training data and its noise too well, to the point where it performs poorly on new, unseen data. The model becomes too complex and memorizes the training data rather than generalizing from it. A tell-tale sign is low training error and high test error. 

2.  **Underfitting:** This occurs when a model is too simple to capture the underlying patterns in the data. It performs poorly on both the training data and new data because it hasn't learned the relationships between the features and the target. A tell-tale sign is high training error and high test error. 

3.  **Bias-Variance Tradeoff:** This is a core concept in machine learning that relates to a model's complexity. **Bias** is the error introduced by approximating a real-world problem with a simplified model (e.g., assuming a linear relationship when the data is non-linear). A high-bias model is too simple and leads to underfitting. **Variance** is the sensitivity of a model to small fluctuations in the training data. A high-variance model is too complex and leads to overfitting. The **tradeoff** is the goal of finding a balance between bias and variance to create a model that generalizes well to new data. Increasing model complexity decreases bias but increases variance, and vice versa.

## Question 4

**What are outliers in a dataset, and list three common techniques for handling them.**

**Answer:**

**Outliers** are data points that are significantly different from the majority of the data. They can be caused by measurement errors, data entry mistakes, or they might represent a genuine rare event.

Three common techniques for handling outliers are:

1.  **Removal:** The simplest method is to remove the outlier from the dataset. This can be effective if the outlier is due to an error and the dataset is large. However, if the outlier is a valid data point, removing it can lead to a loss of valuable information.

2.  **Imputation/Replacement:** Outliers can be replaced with a more representative value, such as the mean, median, or a value based on a statistical model. The median is often preferred for skewed data as it is less sensitive to extreme values.

3.  **Transformation:** A transformation, such as a log or square root transformation, can be applied to the data. This can help reduce the skewness of the distribution and bring the outlier values closer to the other data points.

## Question 5

**Explain the process of handling missing values and mention one imputation technique for numerical and one for categorical data.**

**Answer:**

The process of handling missing values is a crucial step in data preprocessing. It involves identifying and then either removing or replacing the missing data. The approach depends on the type of data and the amount of missingness. A common process is:

1.  **Identification:** First, you identify which columns contain missing values and how many are missing in each. This can be done by checking for `NaN` or other null indicators.
2.  **Analysis:** Analyze the missing data pattern. Is the data missing at random? Is it related to other features? This analysis helps decide the best strategy.
3.  **Imputation or Deletion:** Based on the analysis, you either remove the rows/columns with missing values or impute (fill in) the missing values.

**Imputation Techniques:**
* **Numerical Data:** **Mean/Median Imputation**. This is a simple but effective technique where missing numerical values are replaced by the mean or median of the non-missing values in that column. The median is often more robust to outliers.
* **Categorical Data:** **Mode Imputation**. This involves replacing missing categorical values with the mode (the most frequent category) of the column. This is useful when the missing data is not related to any other feature.

## Question 7

**Implement one-hot encoding using pandas for the following list of colors: `['Red', 'Green', 'Blue', 'Green', 'Red']`. Print the resulting dataframe.**

```python
import pandas as pd

# The list of colors
colors = ['Red', 'Green', 'Blue', 'Green', 'Red']

# Create a Pandas DataFrame from the list
df = pd.DataFrame({'color': colors})

# Apply one-hot encoding using pd.get_dummies()
one_hot_encoded_df = pd.get_dummies(df, columns=['color'], prefix='', prefix_sep='')

# Print the resulting DataFrame
print("Original DataFrame:")
print(df)
print("\nOne-Hot Encoded DataFrame:")
print(one_hot_encoded_df)
```

## Question 8

**Write a Python script to: generate 1000 samples from a normal distribution, introduce 50 random missing values, fill missing values with the column mean, and plot a histogram before and after imputation.**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set a random seed for reproducibility
np.random.seed(42)

# 1. Generate 1000 samples from a normal distribution
data = np.random.normal(loc=10, scale=2, size=1000)

# 2. Introduce 50 random missing values
missing_indices = np.random.choice(range(1000), size=50, replace=False)
data_with_missing = data.copy()
data_with_missing[missing_indices] = np.nan

# 3. Fill missing values with the column mean
mean_value = np.nanmean(data_with_missing)
data_imputed = np.where(np.isnan(data_with_missing), mean_value, data_with_missing)

# 4. Plot histograms before and after imputation
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

sns.histplot(data, bins=30, kde=True, ax=axes[0], color='skyblue')
axes[0].set_title('Original Data Distribution')
axes[0].set_xlabel('Value')
axes[0].set_ylabel('Frequency')
axes[0].grid(axis='y', alpha=0.75)

sns.histplot(data_imputed, bins=30, kde=True, ax=axes[1], color='lightgreen')
axes[1].set_title('Data Distribution After Mean Imputation')
axes[1].set_xlabel('Value')
axes[1].set_ylabel('Frequency')
axes[1].grid(axis='y', alpha=0.75)

plt.tight_layout()
plt.show()

# Print a summary
print("Mean of original data: {:.4f}".format(np.mean(data)))
print("Mean of imputed data: {:.4f}".format(np.mean(data_imputed)))
```

## Question 9

**Implement Min-Max scaling on the following list of numbers `[2, 5, 10, 15, 20]` using `sklearn.preprocessing.MinMaxScaler`. Print the scaled array.**

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# The list of numbers
data = np.array([2, 5, 10, 15, 20]).reshape(-1, 1)

# Create the MinMaxScaler object
scaler = MinMaxScaler()

# Fit the scaler to the data and transform the data
scaled_data = scaler.fit_transform(data)

# Print the scaled array
print("Original data:\n", data.flatten())
print("\nScaled data:\n", scaled_data.flatten())

# Verify the formula: X_scaled = (X - X_min) / (X_max - X_min)
x_min = np.min(data)
x_max = np.max(data)
print("\nVerification using formula:")
for x_val in data.flatten():
    x_scaled = (x_val - x_min) / (x_max - x_min)
    print(f"Scaled value for {x_val}: {x_scaled:.4f}")
```
